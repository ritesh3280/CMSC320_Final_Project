```python
import pandas as pd

df = pd.read_csv('Earthquake_1yr_data.csv')
df
```

```python
df.columns
```

```python
df['time'] = pd.to_datetime(df['time'])
df['date'] = df['time'].dt.date
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
df['depth'] = pd.to_numeric(df['depth'], errors='coerce')
df['mag'] = pd.to_numeric(df['mag'], errors='coerce')
df = df.drop(['type','updated','status','id','time'],axis=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Plot 1: Histogram of earthquake magnitudes
plt.figure(figsize=(8, 6))
plt.hist(df['mag'].dropna(), bins=30, edgecolor='black')
plt.title('Earthquake Magnitudes Distribution')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Box plot for earthquake depth to identify outliers
plt.figure(figsize=(8, 6))
plt.boxplot(df['depth'].dropna(), vert=False)

plt.title('Earthquake Depth')
plt.xlabel('Depth in km')
plt.show()
```

```python
from scipy.stats import pearsonr


plt.figure(figsize=(8, 6))
plt.scatter(df['depth'], df['mag'], alpha=0.3)
plt.title('Magnitude vs. Depth')
plt.xlabel('Depth in km')
plt.ylabel('Magnitude')
plt.show()

# Prepare data: drop missing values for depth and magnitude
valid_data = df[['depth', 'mag']].dropna()

# Perform Pearson correlation test
corr_coef, p_value = pearsonr(valid_data['depth'], valid_data['mag'])
print(f"\nPearson correlation coefficient: {corr_coef:.2f}, p-value: {p_value:.4f}")
```

## Hypothesis Testing  

- **Null Hypothesis (H₀):** There is no correlation between earthquake depth and magnitude.  
- **Alternative Hypothesis (H₁):** There is a significant correlation between earthquake depth and magnitude.  

# Interpretation of Results

- **Pearson correlation coefficient (r) = -0.02** → Almost no correlation between earthquake depth and magnitude.  
- **p-value = 0.1974** → Greater than 0.05, so the correlation is **not statistically significant**.  
- We **fail to reject the null hypothesis**, which means that observed relationship is likely due to chance.  

### Conclusion:  
There is **no strong evidence** that earthquake depth affects magnitude.

```python
# Count earthquakes by state
state_counts = df['state_full'].value_counts().reset_index()
state_counts.columns = ['state', 'count']

# Convert state names to 2-letter codes for the choropleth
us_state_abbrev = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}
state_counts['state_code'] = state_counts['state'].map(us_state_abbrev)

# Implement K-means clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import matplotlib.pyplot as plt

# Standardize the counts
scaler = StandardScaler()
state_counts["count_scaled"] = scaler.fit_transform(state_counts[["count"]])

# Elbow method
inertia = []
K = range(1, 8)
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(state_counts[["count_scaled"]])
    inertia.append(km.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, "o-")
plt.xlabel("k (number of clusters)")
plt.ylabel("Inertia")
plt.title("Elbow Method for K-Means on Earthquake Counts")
plt.show()

# Implement K-means with chosen k
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
state_counts["cluster"] = kmeans.fit_predict(state_counts[["count_scaled"]])

# Create choropleth map
fig = px.choropleth(
    state_counts,
    locations="state",
    locationmode="USA-states",
    color="cluster",
    scope="usa",
    title="Clusters of U.S. States by Seismic Activity"
)
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
fig.show()
```

```python
# Feature selection and preparation
features = df[['mag', 'depth', 'state_full']]
state_stats = features.groupby('state_full').agg({
    'mag': ['count', 'mean'],
    'depth': 'mean'
}).reset_index()

state_stats.columns = ['state', 'quake_count', 'avg_magnitude', 'avg_depth']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(state_stats[['quake_count', 'avg_magnitude', 'avg_depth']])

# Perform clustering
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
state_stats['cluster'] = kmeans.fit_predict(X_scaled)

# Get cluster centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
centroid_df = pd.DataFrame(centroids, 
                          columns=['quake_count', 'avg_magnitude', 'avg_depth'])
print("Cluster centers:")
print(centroid_df)

# Visualization
plt.figure(figsize=(12,8))
scatter = plt.scatter(
    state_stats['avg_magnitude'], 
    state_stats['avg_depth'],
    c=state_stats['cluster'], 
    s=state_stats['quake_count']/10, 
    cmap='tab10', 
    edgecolor='k'
)

for _, row in state_stats.iterrows():
    plt.annotate(row['state'], 
                (row['avg_magnitude'], row['avg_depth']),
                xytext=(5,5), textcoords='offset points')

plt.xlabel("Average Magnitude")
plt.ylabel("Average Depth")
plt.title("State Clustering by Earthquake Characteristics")
plt.colorbar(scatter, label="Cluster")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. LOAD & PREPARE
# df is your DataFrame with a 'time' column
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')

# 2. AGGREGATE TO MONTHLY COUNTS
monthly_counts = df.resample('M').size().rename('quake_count')

# 3. PLOT THE TIME SERIES
plt.figure(figsize=(10,5))
plt.plot(monthly_counts.index, monthly_counts, label='Monthly Count')
plt.xlabel('Date')
plt.ylabel('Number of Earthquakes')
plt.title('Monthly Earthquake Frequency')
plt.grid(True)

# 4. FIT A LINEAR TREND
# turn each month into an integer index 0,1,2,...
X = np.arange(len(monthly_counts)).reshape(-1,1)
y = monthly_counts.values
model = LinearRegression().fit(X, y)
trend = model.predict(X)

plt.plot(monthly_counts.index, trend, color='red', linewidth=2, label='Linear Trend')
plt.legend()
plt.show()

# 5. REPORT THE TREND
slope = model.coef_[0]
intercept = model.intercept_
print(f"Trend: quake_count ≈ {slope:.3f}·month_index + {intercept:.1f}")
if slope > 0:
    print("→ Positive slope indicates an overall increase in frequency over time.")
else:
    print("→ Non‐positive slope suggests no clear increase.")
```
